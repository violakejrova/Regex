```python
valid_email = (r'^[a-z0-9._]*@(gmail.com|seznam.cz)$')

email1 = 'vi.olak3jrova@gmail.com'.lower()
email2 = 'ViOlAk.e.j.rova666@seznam.c'.lower()
email3 = '.....@GMAIL.com'.lower()

x,y,z = re.match(valid_email,email1), re.match(valid_email,email2),re.match(valid_email,email3)

emails = [x, y, z]

for i, result in enumerate(emails):
    if result is None:
        print(f"email {i+1} není validní")
    else:
        print(f"email {i+1} je v pořádku")
```

```python
valid_email = (r'^[a-z0-9._]*@(gmail\.com|seznam\.cz)$')
valid_telefon = (r'^(\+420)?[0-9]{9}$')
valid_web = (r'^(https://)(www\.)?[a-zA-Z0-9]*\.(com|cz|eu|org)/?$')

maily, cisla, weby, jine = [], [], [], []

soubor = open('soubor.txt', 'r').read().split('\n')

for x in soubor:
    if re.match(valid_email, x):
        maily.append(x)
    elif re.match(valid_telefon, x):
        cisla.append(x)
    elif re.match(valid_web, x):
        weby.append(x)
    else:
        jine.append(x)

print(f"'maily: {maily}'\n' telefonní čísla:', {cisla} '\n' weby: {weby}'\n' jiné: {jine}",)
```
